```python
#IMPORTING DATASET
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
data = pd.read_csv("task-02.csv")
data.head(15)

```

```python
print("Description of the data",data.describe())
print("Statistics of the data",data.info())
```

```python
data.drop_duplicates()
```

```python
missing_values = data.isnull().any()  # or df.isna()

print(missing_values)
```

```python
# Data Cleaning
# Fill missing Age values with the median
data['Age'].fillna(data['Age'].median(), inplace=True)

# Create a new feature indicating whether Cabin information is available
data['CabinAvailable'] = data['Cabin'].notna().astype(int)

# Drop the original Cabin column
data.drop(columns=['Cabin'], inplace=True)

# Fill missing Embarked values with the most common value
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)

# Verify if all missing values have been handled
missing_values_after = data.isnull().sum()
print("Missing values after cleaning:\n", missing_values_after)

```

```python
# Plot distributions for various columns
fig, axes = plt.subplots(3, 3, figsize=(15, 15))
sns.countplot(data=data, x='Survived', ax=axes[0, 0]).set_title('Survived')
sns.countplot(data=data, x='Pclass', ax=axes[0, 1]).set_title('Pclass')
sns.countplot(data=data, x='Sex', ax=axes[0, 2]).set_title('Sex')
sns.countplot(data=data, x='SibSp', ax=axes[1, 1]).set_title('SibSp')
sns.countplot(data=data, x='Parch', ax=axes[1, 2]).set_title('Parch')
sns.countplot(data=data, x='Embarked', ax=axes[2, 1]).set_title('Embarked')
sns.countplot(data=data, x='CabinAvailable', ax=axes[2, 2]).set_title('Cabin Available')
sns.histplot(data=data, x='Age', kde=True, ax=axes[1, 0]).set_title('Age')
sns.histplot(data=data, x='Fare', kde=True, ax=axes[2, 0]).set_title('Fare')
plt.tight_layout()
plt.show()
```

```python
#Barplot
sns.barplot(data=data, x='Sex', y='Survived')
plt.title('Survival Rate by Sex')
plt.show()

# Survival rate by passenger class
sns.barplot(data=data, x='Pclass', y='Survived')
plt.title('Survival Rate by Passenger Class')
plt.show()

# Survival rate by embarkation point
sns.barplot(data=data, x='Embarked', y='Survived')
plt.title('Survival Rate by Embarkation Point')
plt.show()
```

```python
# Multivariate Analysis
# Pairplot to see interactions between different features
sns.pairplot(data[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']])
plt.show()
```

```python
#Pie chart
def plot_pie_chart(data, column, title):
    plt.figure(figsize=(8, 8))
    data[column].value_counts().plot.pie(autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
    plt.title(title)
    plt.ylabel('')  # Hide y-label
    plt.show()

# Plot pie charts for various columns

plot_pie_chart(data, 'Survived', 'Survival Distribution')
```

```python
plot_pie_chart(data, 'Pclass', 'Passenger Class Distribution')
```

```python
plot_pie_chart(data, 'Sex', 'Gender Distribution')
```

```python
plot_pie_chart(data, 'Embarked', 'Embarkation Point Distribution')
```

```python
#kdeplot
plt.figure(figsize=(12, 6))
sns.kdeplot(data['Age'], shade=True)
plt.title('Age Distribution')
plt.show()

plt.figure(figsize=(12, 6))
sns.kdeplot(data['Fare'], shade=True)
plt.title('Fare Distribution')
plt.show()


```

```python
# Scatter Plots
plt.figure(figsize=(12, 6))
sns.scatterplot(data=data, x='Age', y='Fare', hue='Survived')
plt.title('Age vs Fare by Survival')
plt.show()

plt.figure(figsize=(12, 6))
sns.scatterplot(data=data, x='SibSp', y='Parch', hue='Survived')
plt.title('SibSp vs Parch by Survival')
plt.show()
```

```python
#Heatplot
numeric_columns = data.select_dtypes(include=['int64', 'float64']).columns

plt.figure(figsize=(10, 8))
corr_matrix = data[numeric_columns].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Boxplot
#Age vs Survived
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Survived', y='Age')
plt.title('Age vs Survived')
plt.show()
# Fare vs Survived
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Survived', y='Fare')
plt.title('Fare vs Survived')
plt.show()

```

```python
# Age vs Pclass
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Pclass', y='Age')
plt.title('Age vs Pclass')
plt.show()

# Fare vs Pclass
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Pclass', y='Fare')
plt.title('Fare vs Pclass')
plt.show()
```
